```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
import seaborn as sns
%matplotlib inline
```

```python
sns.set_style("whitegrid")
```

```python
fig = plt.figure()
ax = plt.axes()

x = np.linspace(0, 10, 100)
ax.plot(x, np.sin(x), color="green", linestyle="--", label="sin")
ax.plot(x, np.cos(x), color="blue", linestyle='-', label="cos")
# ax.set_xlim(-1, 11)
# ax.set_ylim(-1.5, 1.5)
# ax.set_title("Sin function")
# ax.set_xlabel("x")

ax.set(xlim=(-1, 11), 
       ylim=(-1.5, 1.5),
       title="functions",
       xlabel="x")

ax.legend()
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
import seaborn as sns
%matplotlib inline
```

```python
sns.set_style("whitegrid")
```

```python
x = np.linspace(0, 10, 30)
fig = plt.figure()
ax = plt.axes()

ax.plot(x, np.sin(x))
ax.scatter(x, np.sin(x), marker='o')
```

```python
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.randn(100)
sizes = 1000 * rng.randn(100)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.3, cmap="viridis")
plt.colorbar()
```

## Let's visualize IRIS dataset

```python
from sklearn.datasets import load_iris
iris = load_iris()
features = iris.data.T

plt.scatter(features[0], features[1], c=iris.target, s=100*features[3], cmap='viridis')
```

```python
fig = plt.figure()
ax = plt.axes()

ax.scatter(features[0], features[1], c=iris.target, s=100*features[3], cmap='viridis')
```
